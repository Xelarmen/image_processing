```python
#Abdulkadir KAYA
#19-701-806
import cv2
import numpy as np
import matplotlib.pyplot as plt

resim = cv2.imread('arac-plakasi.jpg')
gray=cv2.cvtColor(resim,cv2.COLOR_BGR2GRAY)
gauss=cv2.GaussianBlur(gray,(3,3),0)
plt.imshow(resim)
plt.title('Orijinal')
```

```python
#Previtt
kernel_Previtt_x=np.array([[-1, 0, 1],
                           [-1, 0, 1],
                           [-1, 0, 1]])
kernel_Previtt_y=np.array([[1, 1, 1],
                           [0, 0, 0],
                           [-1,-1,-1]])
prewittx=cv2.filter2D(gauss,-1,kernel_Previtt_x)
prewitty=cv2.filter2D(gauss,-1,kernel_Previtt_y)
plt.imshow(prewittx + prewitty,cmap='gray')
plt.title('Prewitt')
```

```python
#Roberts
kernel_Roberts_x = np.array([[1, 0],
                             [0, -1]])
kernel_Roberts_y = np.array([[0, -1],
                             [1, 0]])
robertsx = cv2.filter2D(gauss, -1, kernel_Roberts_x)
robertsy = cv2.filter2D(gauss, -1, kernel_Roberts_y)
plt.imshow(robertsx + robertsy,cmap='gray')
plt.title('Roberts')
```

```python
#Sobel
kernel_Sobel_x = np.array([[-1, 0, -1],
                           [-2, 0, 2],
                           [-1, 0, 1]])
kernel_Sobel_y = np.array([[1, 2, 1],
                           [0, 0, 0],
                           [-1,-2,-1]])
sobelx = cv2.filter2D(gauss, -1, kernel_Sobel_x)
sobely = cv2.filter2D(gauss, -1, kernel_Sobel_y)
plt.imshow(sobelx + sobely,cmap='gray')
plt.title('Sobel')
```
